```python
iris = px.data.iris()
iris.head()
```

```python
tips = px.data.tips()
tips.head()
```

```python
fig=px.scatter(tips,x='total_bill',y='tip',facet_col='sex',facet_row='time',color='smoker')
fig.update_layout(height=500)
fig.show()
```

```python
fig=px.scatter(tips,x='total_bill',y='tip',trendline='ols')
fig.show()
```

##### Histogram


```python
fig = px.histogram(tips,x='total_bill'b)
fig.show()
```

```python
fig = px.histogram(tips,x='total_bill',nbins=20)
fig.show()
```

##### Box plot


```python
fig = px.box(tips,y='total_bill')
fig.show()
```

```python
fig = px.box(tips,x='time',y='total_bill',points='all')
fig.show()
```

```python
fig = px.box(tips,x='time',y='total_bill',points='outliers',color='smoker')
fig.show()
```

```python
fig = px.box(tips,x='time',y='total_bill',points='outliers',color='smoker',notched=True,title='Outliers in the total bill')
fig.show()
```

```python
# get the points and not the box
fig = px.strip(tips,x='day',y='tip')
fig.show()
```

###### Violin chart


```python
fig = px.violin(tips,y='total_bill')
fig.show()
```

```python
fig = px.violin(tips,y='total_bill',points='all',box=True)
fig.show()
```

```python
fig = px.violin(tips,y='tip',color='time',violinmode='overlay',hover_data=tips.columns)
fig.show()
```

```python
fig = px.violin(tips,y='total_bill')
fig.update_traces(quartilemethod='exclusive')
fig.show()
```
